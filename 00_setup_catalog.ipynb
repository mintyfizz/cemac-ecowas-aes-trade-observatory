```python
# creating a catalog for my project 

spark.sql("CREATE CATALOG IF NOT EXISTS cemac_ecowas_aes_trade")
```

```python
# now im going to use the catalog to create schemas

spark.sql("USE CATALOG cemac_ecowas_aes_trade")

schemas = ["bronze", "silver", "gold", "audit"]

for schema in schemas:
    spark.sql(f"CREATE SCHEMA IF NOT EXISTS {schema}")
    print(f"{str(schema).upper()} schema  created successfully.")

print("All schemas created successfully.")
```

```python
spark.sql("SHOW SCHEMAS IN cemac_ecowas_aes_trade").show()
```

```python
spark.sql("USE CATALOG cemac_ecowas_aes_trade")

spark.sql("USE SCHEMA bronze")
spark.sql("CREATE VOLUME IF NOT EXISTS raw_landing")
print("Volume created at: /Volumes/cemac_ecowas_aes_trade/bronze/raw_landing")
```
